```python
import nibabel as nib
from nilearn import plotting, image
```

## imagesTr (Training images)

```python
tr_image = nib.load('../data/brats/imagesTr/BRATS_001.nii.gz')
# tr_image is in 4D form
# in its last dimension, it contains:
# Index 0: The T1 scan
# Index 1: The T1-contrast (T1c) scan
# Index 2: The T2 scan
# Index 3: The FLAIR scan

# lets try the flair scan
tr_image_flair = image.index_img(tr_image, 3)

view = plotting.view_img(tr_image_flair,
                         bg_img=False,
                         title="Interactive FLAIR Scan in 3D")

view    
```

## imagesTs (Testing images)

```python
ts_image = nib.load('../data/brats/imagesTs/BRATS_485.nii.gz')

# lets try dimension 0 this time (the T1 scan)
ts_image_flair = image.index_img(ts_image, 0) # we can change the dimension to 1, 2 or 3

view = plotting.view_img(ts_image_flair,
                        #  bg_img=False,
                         title="Interactive T1 Scan in 3D")

view    
```
